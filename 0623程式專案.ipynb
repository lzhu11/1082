```python
#成人吸菸比率
import matplotlib.pyplot as plt
import pandas as pd
tdata=pd.read_csv("18歲以上人口目前吸菸率 (1).csv",encoding='big5')
spots=tdata[tdata["類別"] == "整體"]
spots[['年度','總計']].plot(kind='line',x='年度',y='總計',legend=False,linewidth=2,color="red",marker="o",ms=8,markerfacecolor="blue",markeredgecolor="red")
plt.vlines(x=2002, ymin=0, ymax=30,linewidth=2,linestyle="-.",color="LightSlateGray")
plt.vlines(x=2006, ymin=0, ymax=30,linewidth=2,linestyle="-.",color="LightSlateGray")
plt.vlines(x=2009, ymin=0, ymax=30,linewidth=2,linestyle="-.",color="LightSlateGray")
plt.vlines(x=2017, ymin=0, ymax=30,linewidth=2,linestyle="-.",color="LightSlateGray")
plt.title('成人吸菸比率',fontsize=15,fontweight="bold")
plt.xlim(2001,2019)
plt.ylim(0,30)
plt.xlabel("年度",fontweight="bold")
plt.ylabel("比率(%)",fontweight="bold")
x=spots["年度"]
y=spots["總計"]
plt.bar(2001.5,11,color="PowDerBlue")
plt.bar(2001.5,9,color="white")
plt.bar(2004,11,color="PowDerBlue")
plt.bar(2004,9,color="white")
plt.bar(2007.3,11,color="PowDerBlue")
plt.bar(2007.3,9,color="white")
plt.bar(2012.4,11,color="PowDerBlue")
plt.bar(2012.4,9,color="white")
plt.bar(2012.6,11,color="PowDerBlue")
plt.bar(2012.6,9,color="white")
plt.bar(2017.6,11,color="PowDerBlue")
plt.bar(2017.6,9,color="white")
plt.bar(2018.4,11,color="PowDerBlue")
plt.bar(2018.4,9,color="white")
for x,y in zip(x,y):
    plt.text(x,y+0.5,"%.1f"%y,ha="center",va="bottom",fontsize=8)
plt.annotate("80",xy=(2001.2,9.5))
plt.annotate("85",xy=(2003.7,9.5))
plt.annotate("90",xy=(2007,9.5))
plt.annotate("100",xy=(2012,9.5))
plt.annotate("120(元)",xy=(2017.1,9.5))
plt.grid()
#plt.savefig('成人吸菸比率.png')
plt.show()
```

import imageio
 
def create_gif(image_list, gif_name):
 
    frames = []
    for image_name in image_list:
        frames.append(imageio.imread(image_name))
    imageio.mimsave(gif_name, frames, 'GIF', duration = 1)
 
    return
