```python
#第四題
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=pd.read_csv('cases_add_Taiwan.csv')
a=data['Date']
b=data['Italy']
c=data['Singapore']
d=data['Austria']
plt.plot(a,b,label='Italy',ls='--')
plt.plot(a,c,label='Singapore',ls=':')
plt.plot(a,d,label='Austria')
plt.title('Covid-19 Daily Reports')
plt.legend()
plt.xlim(43,54)
plt.ylim(0,2500)
plt.xlabel("Date")
plt.ylabel("cases")
plt.show()
```
